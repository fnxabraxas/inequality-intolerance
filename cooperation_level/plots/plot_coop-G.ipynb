```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.patches as patches

yv = [0.1, 0.3, 0.5, 0.7, 0.9]
strgv =  [2453] #[405, 469] # [3477, 3541,  1437,  1501 ] #[1437,2517]

epsT=0.3
```

```python
plt.rc('xtick', labelsize=15)    # fontsize of the tick labels
plt.rc('ytick', labelsize=15)    # fontsize of the tick labels
mpl.rcParams['axes.titlepad'] = 20 

cb = "royalblue"
ab = 0.1

epsT=0.5

dir_R_TT = '../broteR_TT_bcte_B/'
dir_T_RT = '../broteT_RT_bcte_B/'

rows=5
yvr = [0.1, 0.3, 0.5, 0.7, 0.9]
for istrg,strg in enumerate(strgv):
       
    fig, axes = plt.subplots(rows,len(yv) , figsize=(20,20)) #, sharex='all', sharey='all')
    fig.subplots_adjust(hspace=0.25, wspace=0.25, left=0.07, right=0.85)
    
    for iy,y in enumerate(yvr):
        ff=1 #(1.-epsT)/(y*(1-y))
        
        coop_ARtoA_TT = np.loadtxt(dir_R_TT + 'coop_G_ARtoA_broteR_TT_epsT_' + str(epsT) + '_y_' + str(y) + '_stg_' + str(strg) + '.dat'); 
        coop_ARtoB_TT = np.loadtxt(dir_R_TT + 'coop_G_ARtoB_broteR_TT_epsT_' + str(epsT) + '_y_' + str(y) + '_stg_' + str(strg) + '.dat') 
        coop_AMtoA_TT = np.loadtxt(dir_R_TT + 'coop_G_AMtoA_broteR_TT_epsT_' + str(epsT) + '_y_' + str(y) + '_stg_' + str(strg) + '.dat')   
        coop_AR_TT = coop_ARtoA_TT.copy(); coop_AR_TT[:,1] = y*coop_ARtoA_TT[:,1] + (1-y)*coop_ARtoB_TT[:,1]  # it is the same for both AR and BR
        coop_AM_TT = coop_ARtoA_TT.copy(); coop_AM_TT[:,1] = y*coop_AMtoA_TT[:,1]
        
        coop_ARtoA_RT = np.loadtxt(dir_T_RT + 'coop_G_ARtoA_broteT_RT_epsT_' + str(epsT) + '_y_' + str(y) + '_stg_' + str(strg) + '.dat') 
        coop_BRtoA_RT = np.loadtxt(dir_T_RT + 'coop_G_BRtoA_broteT_RT_epsT_' + str(epsT) + '_y_' + str(y) + '_stg_' + str(strg) + '.dat') 
        coop_BRtoB_RT = np.loadtxt(dir_T_RT + 'coop_G_BRtoB_broteT_RT_epsT_' + str(epsT) + '_y_' + str(y) + '_stg_' + str(strg) + '.dat') 
        coop_AR_RT = coop_ARtoA_RT.copy(); coop_AR_RT[:,1] = y*coop_ARtoA_RT[:,1]
        coop_BR_RT = coop_ARtoA_RT.copy(); coop_BR_RT[:,1] = y*coop_BRtoA_RT[:,1] + (1-y)*coop_BRtoB_RT[:,1]
        coop_R_RT = coop_ARtoA_RT.copy(); coop_R_RT[:,1] = y*coop_AR_RT[:,1] + (1-y)*coop_BR_RT[:,1]
        
        axes[0,iy].set_title('$y=$'+ str(y),fontsize=17)
        for irow in range(rows): 
            axes[irow,iy].axvline(x=0, color='darkgray', ls='-', lw='1.5')

        irow=0 # before invasion
        axes[irow,iy].plot(coop_ARtoA_TT[:,0]*ff, coop_ARtoA_TT[:,1],c='Red',ls='-',lw=1.5, label=r'A/B/AB$\rightarrow$A (TT)')
        axes[irow,iy].plot(coop_ARtoB_TT[:,0]*ff, coop_ARtoB_TT[:,1],c='violet',ls='-',lw=1.5, label=r'A/B/AB$\rightarrow$B (TT)')
        axes[irow,iy].plot(coop_AR_TT[:,0]*ff, coop_AR_TT[:,1],c='Red',ls='-',lw=4, label='A/B/AB (TT)')
        axes[irow,iy].plot(coop_AMtoA_TT[:,0]*ff, coop_AMtoA_TT[:,1],c='Orange',ls='-',lw=1.5, label=r'A$_M\rightarrow$A (TT)')
        axes[irow,iy].scatter([-1], [-1],edgecolors='Orange',marker='o',s=80,facecolors='none', label=r'A$_M\rightarrow$B (TT)')
        axes[irow,iy].plot(coop_AM_TT[:,0]*ff, coop_AM_TT[:,1],c='Orange',ls='-',lw=3, label='A$_M$ (TT)')
        
        irow=1 # after invasion
        axes[irow,iy].plot(coop_ARtoA_RT[:,0]*ff, coop_ARtoA_RT[:,1],c='Blue',ls='-',lw=1.5, label=r'A$\rightarrow$A (IT)')
        axes[irow,iy].scatter([-1], [-1],edgecolors='Green',marker='o',s=80,facecolors='none', label=r'A$\rightarrow$B (IT)')
        axes[irow,iy].plot(coop_AR_RT[:,0]*ff, coop_AR_RT[:,1],c='Blue',ls='-',lw=3, label='A (IT)')
        axes[irow,iy].plot(coop_BRtoA_RT[:,0]*ff, coop_BRtoA_RT[:,1],c='Blue',ls='--',lw=1.5, label=r'B$\rightarrow$A (IT)')
        axes[irow,iy].plot(coop_BRtoB_RT[:,0]*ff, coop_BRtoB_RT[:,1],c='Green',ls='--',lw=1.5, label=r'B$\rightarrow$B (IT)')
        axes[irow,iy].plot(coop_BR_RT[:,0]*ff, coop_BR_RT[:,1],c='Blue',ls='--',lw=3, label='B (IT)')
        axes[irow,iy].plot(coop_R_RT[:,0]*ff, coop_R_RT[:,1],c='DarkBlue',ls='-',lw=4, label='AB (IT)')
        
        irow=2 # in-group cooperation
        axes[irow,iy].plot(coop_ARtoA_TT[:,0]*ff, coop_ARtoA_TT[:,1],c='Red',ls='-',lw=1.5, label=r'A$\rightarrow$A (TT)')
        axes[irow,iy].plot(coop_ARtoB_TT[:,0]*ff, coop_ARtoB_TT[:,1],c='Red',ls='--',lw=1.5, label=r'B$\rightarrow$B (TT)')
        axes[irow,iy].plot(coop_AMtoA_TT[:,0]*ff, coop_AMtoA_TT[:,1],c='Orange',ls='-',lw=2, label=r'A$_M\rightarrow$A (TT)')
        axes[irow,iy].plot(coop_ARtoA_RT[:,0]*ff, coop_ARtoA_RT[:,1],c='Blue',ls='-',lw=1.5, label=r'A$\rightarrow$A (IT)')
        axes[irow,iy].plot(coop_BRtoB_RT[:,0]*ff, coop_BRtoB_RT[:,1],c='Blue',ls='--',lw=1.5, label=r'B$\rightarrow$B (IT)')
        
        irow=3 # out-group cooperation     
        axes[irow,iy].plot(coop_ARtoB_TT[:,0]*ff, coop_ARtoB_TT[:,1],c='Red',ls='-',lw=1.5, label=r'A$\rightarrow$B (TT)')
        axes[irow,iy].plot(coop_ARtoA_TT[:,0]*ff, coop_ARtoA_TT[:,1],c='Red',ls='--',lw=1.5, label=r'B$\rightarrow$A (TT)')
        axes[irow,iy].scatter([-1], [-1],edgecolors='blue',marker='o',s=80,facecolors='none', label=r'A$\rightarrow$B (IT)')
        axes[irow,iy].plot(coop_BRtoA_RT[:,0]*ff, coop_BRtoA_RT[:,1],c='blue',ls='--',lw=1.5, label=r'B$\rightarrow$A (IT)')
        
        irow=4 # total
        axes[irow,iy].plot(coop_AR_TT[:,0]*ff, coop_AR_TT[:,1],c='Red',ls='-',lw=4, label='A/B/AB (TT)')
        axes[irow,iy].plot(coop_AM_TT[:,0]*ff, coop_AM_TT[:,1],c='Orange',ls='-',lw=1.5, label='A$_M$ (TT)')
        axes[irow,iy].plot(coop_AR_RT[:,0]*ff, coop_AR_RT[:,1],c='Blue',ls='-',lw=1.5, label='A (IT)')
        axes[irow,iy].plot(coop_BR_RT[:,0]*ff, coop_BR_RT[:,1],c='Blue',ls='--',lw=1.5, label='B (IT)')
        axes[irow,iy].plot(coop_R_RT[:,0]*ff, coop_R_RT[:,1],c='Blue',ls='-',lw=4, label='AB (IT)') 

        for irow in range(rows): 
            #axes[irow,iy].set_yticks([0.2,0.4,0.6,0.8])
            #axes[irow,iy].set_xticks([-0.5, -0.3, -0.1,  0.1,0.3, 0.5])
            #axes[irow,iy].axvline(x=epsv_TT[iepsA],c='black',ls='--')
            axes[irow,iy].set_ylim([0.18,1]); axes[irow,iy].set_xlim([np.min(coop_R_RT[:,0])*ff,np.max(coop_R_RT[:,0])*ff])
            
#        # by hand:
#        for irow in range(rows):
#            if (strg==1437):
#                axes[irow,0].add_patch(patches.Rectangle((0,0),0.4902-0,1, alpha=ab,facecolor=cb))
#                axes[irow,0].add_patch(patches.Rectangle((0.7646,0),0.9802-0.7646,1, alpha=ab,facecolor=cb)) 
#                axes[irow,1].add_patch(patches.Rectangle((0,0),0.451-0,1, alpha=ab,facecolor=cb))
#                axes[irow,2].add_patch(patches.Rectangle((0,0),0.3922-0,1, alpha=ab,facecolor=cb))
#                #axes[irow,3].add_patch(patches.Rectangle((0,0),0.2844-0,1, alpha=ab,facecolor=cb))
#            elif (strg==2517):
#                axes[irow,0].add_patch(patches.Rectangle((0,0),0.3432-0,1, alpha=ab,facecolor=cb))
#                axes[irow,0].add_patch(patches.Rectangle((0.6568,0),1-0.6568,1, alpha=ab,facecolor=cb))
#                axes[irow,1].add_patch(patches.Rectangle((0,0),0.1374-0,1, alpha=ab,facecolor=cb))
#                axes[irow,1].add_patch(patches.Rectangle((0.8724,0),1-0.8724,1, alpha=ab,facecolor=cb))
                
    #fig.text(0.06, 0.5, '$\epsilon_A$', va='center', rotation='vertical',size=25)
    #fig.text(0.5, 0.08, '$\epsilon_B$', va='center',size=25)

    for irow in range(rows):
        axes[irow,0].set_xticks([-0.1,  0,0.1]) 
        axes[irow,1].set_xticks([-0.2, 0,0.2]) 
        axes[irow,2].set_xticks([-0.4, -0.2, 0,  0.2,0.4]) 
        axes[irow,3].set_xticks([-0.2, 0,0.2])
        axes[irow,4].set_xticks([-0.1,  0,0.1]) 
        
 
    
    fsl = 15
    leg = axes[0,4].legend(fontsize=fsl, loc='upper left',bbox_to_anchor=(1.1,1.02), title_fontsize=fsl,frameon=False,  title=r'$\bf{Before\ invasion}$'); leg._legend_box.align = "left"
    leg = axes[1,4].legend(fontsize=fsl, loc='upper left',bbox_to_anchor=(1.1,1.02), title_fontsize=fsl,frameon=False, title=r'$\bf{After\ invasion}$'); leg._legend_box.align = "left"
    leg = axes[2,4].legend(fontsize=fsl, loc='upper left',bbox_to_anchor=(1.1,1.02), title_fontsize=fsl,frameon=False, title=r'$\bf{Ingroup}$'); leg._legend_box.align = "left"
    leg = axes[3,4].legend(fontsize=fsl, loc='upper left',bbox_to_anchor=(1.1,1.02), title_fontsize=fsl,frameon=False, title=r'$\bf{Outgroup}$'); leg._legend_box.align = "left"
    leg = axes[4,4].legend(fontsize=fsl, loc='upper left',bbox_to_anchor=(1.1,1.02), title_fontsize=fsl,frameon=False, title=r'$\bf{Total}$'); leg._legend_box.align = "left"
    #[axes[i,4].axis('off') for i in range(5)]
    
            
    fig.text(0.01, 0.5, 'Intention of cooperation', va='center', rotation='vertical',size=25)
    fig.text(0.45, 0.08, 'G', va='center',size=25)
        
    name = 'coop_all_G_epsT_' + str(epsT) + '_' + str(strg)
    fig.savefig(name + '.svg')
    fig.savefig(name + '.jpg')
    fig.savefig(name + '.pdf')
    plt.close()
```
